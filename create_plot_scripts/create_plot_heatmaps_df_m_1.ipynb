```python
models = ['ExtraTreeClassifier', 'BaggingClassifier', 'AdaBoostClassifier',
 'KNeighborsClassifier', 'DecisionTreeClassifier', 'GaussianNB',
 'ExtraTreesClassifier', 'LabelPropagation', 'RandomForestClassifier',
 'NuSVC', 'LabelSpreading', 'SVC', 'XGBClassifier',
 'QuadraticDiscriminantAnalysis', 'LGBMClassifier',
 'LinearDiscriminantAnalysis', 'CalibratedClassifierCV', 'DummyClassifier',
 'LinearSVC', 'SGDClassifier', 'RidgeClassifier', 'RidgeClassifierCV',
 'LogisticRegression', 'PassiveAggressiveClassifier', 'Perceptron',
 'NearestCentroid', 'BernoulliNB']

models = ['ExtraTreeClassifier', 'BaggingClassifier', 'AdaBoostClassifier',
 'KNeighborsClassifier', 'DecisionTreeClassifier', 'GaussianNB',
 'ExtraTreesClassifier', 'LabelPropagation', 'RandomForestClassifier']

scenarios = ["Base", "Base + Synthetic data", "Base + Synthetic data (FP Dataset)", "Casing", "Missing Value", "Floating Point Number As String", "Suspect Sign"]

models_data_f1_score = np.load("./outputComparisons/Models/Heatmaps_df_m_1/f1_score.npy")[0:9]

models_data_f1_score = models_data_f1_score.round(2)
models_data_bal_accuracy = np.load("./outputComparisons/Models/Heatmaps_df_m_1/bal_accuracy.npy")
#models_data_f1_score = np.load("./outputComparisons/Models/Heatmaps_df_m_1/f1_score.npy")
models_data_time = np.load("./outputComparisons/Models/Heatmaps_df_m_1/time.npy")

fig, heatmap = plt.subplots()
imshow = heatmap.imshow(models_data_f1_score)
heatmap.set_xticks(np.arange(len(scenarios)), labels=scenarios)
heatmap.set_yticks(np.arange(len(models)), labels=models)
plt.setp(heatmap.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
for i in range(len(models)):
    for j in range(len(scenarios)):
        text = heatmap.text(j, i, models_data_f1_score[i, j],
                       ha="center", va="center", color="w")
        

heatmap.set_title("Heatmap of F1 Score averages of classification models, \ntrained on different types of \"Firefighter_Promotion_Exam_Scores\" \nwith categorical feature [a]", pad =20)
plt.show()
```
