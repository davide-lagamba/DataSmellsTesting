```python
#tutorial https://medium.com/omics-diary/how-to-use-the-lazy-predict-library-to-select-the-best-machine-learning-model-65378bf4568e
!pip install lazypredict
import pandas as pd
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
!mkdir "outputLazyPredict"
!mkdir "outputLazyPredict/BaseSynthetic"

output_path_l_1 = "outputLazyPredict/BaseSynthetic/df_l_1.csv"
output_path_l_2 = "outputLazyPredict/BaseSynthetic/df_l_2.csv"
output_path_l_3 = "outputLazyPredict/BaseSynthetic/df_l_3.csv"
output_path_m_1 = "outputLazyPredict/BaseSynthetic/df_m_1.csv"
output_path_m_1_FP = "outputLazyPredict/BaseSynthetic/df_m_1_FP.csv"
output_path_m_2 = "outputLazyPredict/BaseSynthetic/df_m_2.csv"
```

```python
df_l_1_path = "./cleaned_datasets/df_l_1.csv"
df_l_2_path = "./cleaned_datasets/df_l_2.csv"
df_l_3_path = "./cleaned_datasets/df_l_3.csv"
df_m_1_path = "./cleaned_datasets/df_m_1.csv"
df_m_2_path = "./cleaned_datasets/df_m_2.csv"
df_l_1 = pd.read_csv(df_l_1_path, index_col=False)
df_l_2 = pd.read_csv(df_l_2_path, index_col=False)
df_l_3 = pd.read_csv(df_l_3_path, index_col=False)
df_m_1 = pd.read_csv(df_m_1_path, index_col=False)
df_m_2 = pd.read_csv(df_m_2_path, index_col=False)
```

```python
ds_df_l_1_path = "./outputDSInjection/BaseSynthetic/df_l_1_SDV.csv"
ds_df_l_2_path = "./outputDSInjection/BaseSynthetic/df_l_2_SDV.csv"
ds_df_l_3_path = "./outputDSInjection/BaseSynthetic/df_l_3_SDV.csv"
ds_df_m_1_path = "./outputDSInjection/BaseSynthetic/df_m_1_SDV.csv"
ds_df_m_1_FP_path = "./outputDSInjection/BaseSynthetic/df_m_1_FP_SDV.csv"
ds_df_m_2_path = "./outputDSInjection/BaseSynthetic/df_m_2_SDV.csv"
ds_df_l_1 = pd.read_csv(ds_df_l_1_path, index_col=False)
ds_df_l_2 = pd.read_csv(ds_df_l_2_path, index_col=False)
ds_df_l_3 = pd.read_csv(ds_df_l_3_path, index_col=False)
ds_df_m_1 = pd.read_csv(ds_df_m_1_path, index_col=False)
ds_df_m_1_FP = pd.read_csv(ds_df_m_1_FP_path, index_col=False)
ds_df_m_2 = pd.read_csv(ds_df_m_2_path, index_col=False)
```

```python
x1 =  df_l_1.drop('income_level', axis=1)
ds_x1 = ds_df_l_1.drop('income_level', axis=1)
y1 = df_l_1['income_level']
ds_y1 = ds_df_l_1['income_level']
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.2, random_state =42)
x1_train = pd.concat([x1_train, ds_x1])
y1_train = pd.concat([y1_train, ds_y1])

x2 =  df_l_2.drop(columns=['readmitted'])
ds_x2 = ds_df_l_2.drop('readmitted', axis=1)
y2 = df_l_2['readmitted']
ds_y2 = ds_df_l_2['readmitted']
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.2, random_state =42)
x2_train = pd.concat([x2_train, ds_x2])
y2_train = pd.concat([y2_train, ds_y2])

x3 =  df_l_3.drop(columns=['class'])
ds_x3 = ds_df_l_3.drop('class', axis=1)
y3 = df_l_3['class']
ds_y3 = ds_df_l_3['class']
x3_train, x3_test, y3_train, y3_test = train_test_split(x3, y3, test_size=0.2, random_state =42)
x3_train = pd.concat([x3_train, ds_x3])
y3_train = pd.concat([y3_train, ds_y3])

x4 =  df_m_1.drop(columns=['promotion'])
ds_x4 = ds_df_m_1.drop('promotion', axis=1)
y4 = df_m_1['promotion']
ds_y4 = ds_df_m_1['promotion']
x4_train, x4_test, y4_train, y4_test = train_test_split(x4, y4, test_size=0.2, random_state =42)
x4_train = pd.concat([x4_train, ds_x4])
y4_train = pd.concat([y4_train, ds_y4])

x4_FP =  df_m_1.drop(columns=['promotion'])
ds_x4_FP = ds_df_m_1_FP.drop('promotion', axis=1)
y4_FP = df_m_1['promotion']
ds_y4_FP = ds_df_m_1_FP['promotion']
x4_train_FP, x4_test_FP, y4_train_FP, y4_test_FP = train_test_split(x4_FP, y4_FP, test_size=0.2, random_state =42)
x4_train_FP = pd.concat([x4_train_FP, ds_x4_FP])
y4_train_FP = pd.concat([y4_train_FP, ds_y4_FP])

x5 =  df_m_2.drop(columns=['Class_attribute'])
ds_x5 = ds_df_m_2.drop('Class_attribute', axis=1)
y5 = df_m_2['Class_attribute']
ds_y5 = ds_df_m_2['Class_attribute']
x5_train, x5_test, y5_train, y5_test = train_test_split(x5, y5, test_size=0.2, random_state =42)
x5_train = pd.concat([x5_train, ds_x5])
y5_train = pd.concat([y5_train, ds_y5])

categorical_x_train_l_1_1 = x1_train.select_dtypes(include="object")
for i in categorical_x_train_l_1_1:
  x1_train[i] = pd.factorize(x1_train[i])[0]
categorical_x_test_l_1_1 = x1_test.select_dtypes(include="object")
for i in categorical_x_test_l_1_1:
  x1_test[i] = pd.factorize(x1_test[i])[0]
if(y1_train.dtype == 'object'):
  y1_train, _ = pd.factorize(y1_train)
if(y1_test.dtype == 'object'):
  y1_test, _ = pd.factorize(y1_test)


categorical_x_train_l_2_1 = x2_train.select_dtypes(include="object")
for i in categorical_x_train_l_2_1:
  x2_train[i] = pd.factorize(x2_train[i])[0]
categorical_x_test_l_2_1 = x2_test.select_dtypes(include="object")
for i in categorical_x_test_l_2_1:
  x2_test[i] = pd.factorize(x2_test[i])[0]
if(y2_train.dtype == 'object'):
  y2_train, _ = pd.factorize(y2_train)
if(y2_test.dtype == 'object'):
  y2_test, _ = pd.factorize(y2_test)


categorical_x_train_l_3_1 = x3_train.select_dtypes(include="object")
for i in categorical_x_train_l_3_1:
  x3_train[i] = pd.factorize(x3_train[i])[0]
categorical_x_test_l_3_1 = x3_test.select_dtypes(include="object")
for i in categorical_x_test_l_3_1:
  x3_test[i] = pd.factorize(x3_test[i])[0]
if(y3_train.dtype == 'object'):
  y3_train, _ = pd.factorize(y3_train)
if(y3_test.dtype == 'object'):
  y3_test, _ = pd.factorize(y3_test)


categorical_x_train_m_1_1 = x4_train.select_dtypes(include="object")
for i in categorical_x_train_m_1_1:
  x4_train[i] = pd.factorize(x4_train[i])[0]
categorical_x_test_m_1_1 = x4_test.select_dtypes(include="object")
for i in categorical_x_test_m_1_1:
  x4_test[i] = pd.factorize(x4_test[i])[0]
if(y4_train.dtype == 'object'):
  y4_train, _ = pd.factorize(y4_train)
if(y4_test.dtype == 'object'):
  y4_test, _ = pd.factorize(y4_test)


categorical_x_train_m_1_1_FP = x4_train_FP.select_dtypes(include="object")
for i in categorical_x_train_m_1_1_FP:
  x4_train_FP[i] = pd.factorize(x4_train_FP[i])[0]
categorical_x_test_m_1_1_FP = x4_test_FP.select_dtypes(include="object")
for i in categorical_x_test_m_1_1_FP:
  x4_test_FP[i] = pd.factorize(x4_test_FP[i])[0]
if(y4_train_FP.dtype == 'object'):
  y4_train_FP, _ = pd.factorize(y4_train_FP)
if(y4_test_FP.dtype == 'object'):
  y4_test_FP, _ = pd.factorize(y4_test_FP)


categorical_x_train_m_2_1 = x5_train.select_dtypes(include="object")
for i in categorical_x_train_m_2_1:
  x5_train[i] = pd.factorize(x5_train[i])[0]
categorical_x_test_m_2_1 = x5_test.select_dtypes(include="object")
for i in categorical_x_test_m_2_1:
  x5_test[i] = pd.factorize(x5_test[i])[0]
if(y5_train.dtype == 'object'):
  y5_train, _ = pd.factorize(y5_train)
if(y5_test.dtype == 'object'):
  y5_test, _ = pd.factorize(y5_test)
```

```python
classifier_1 = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_1,predictions_1 = classifier_1.fit(x1_train, x1_test, y1_train, y1_test)
models_1.to_csv(output_path_l_1)
models_1
```

```python
classifier_2 = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_2,predictions_2 = classifier_2.fit(x2_train, x2_test, y2_train, y2_test)
models_2.to_csv(output_path_l_2)
models_2
```

```python
classifier_3 = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_3,predictions_3= classifier_3.fit(x3_train, x3_test, y3_train, y3_test)
models_3.to_csv(output_path_l_3)
predictions_3
```

```python
classifier_4 = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_4,predictions_4 = classifier_4.fit(x4_train, x4_test, y4_train, y4_test)
models_4.to_csv(output_path_m_1)
models_4
```
